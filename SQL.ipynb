```python
import pandas as pd
import numpy as np
```

```python
import sqlite3
# Connect to the SQLite database
conn = sqlite3.connect('im.db')

# Query the list of tables in the database
query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, conn)

print(tables)
```

```python
pd.read_sql('''SELECT * FROM movie_basics LIMIT 2''',conn)
```

```python
pd.read_sql('''SELECT * FROM movie_ratings LIMIT 2''',conn)
```

```python
# joining the movie_basics table and the movie_ratings table

pd.read_sql('''SELECT *
                FROM movie_basics
                JOIN movie_ratings
                USING(movie_id)
                ORDER BY numvotes DESC
                 LIMIT 10''',conn)
```

- Movie with the title inceptions received the highest number of votes at 1841066 votes

```python

pd.read_sql('''SELECT COUNT(*)
                FROM movie_basics
                JOIN movie_ratings
                USING(movie_id)
                WHERE averagerating = 10.0
                 LIMIT 10''',conn)
```

```python
pd.read_sql('''SELECT genres, averagerating
                FROM movie_basics
                JOIN movie_ratings
                USING(movie_id)
               ORDER BY averagerating DESC
                 LIMIT 17''',conn)
```

- Genre documentary has the highest count of the movies with the highest rating therefore a good genre to venture into for a company.
- other genres that has a rating of 10 are: comedy,Drama,crime,History, and Adventure.

```python
pd.read_sql('''SELECT genres, numvotes
                FROM movie_basics
                JOIN movie_ratings
                USING(movie_id)
               ORDER BY numvotes DESC
                 LIMIT 1''',conn)
```

- genres Action, Adventure and Sci-Fi are the ones with the highest number of votes
